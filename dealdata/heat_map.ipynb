```python
import pandas as pd
# 读取player 的数据
player = pd.read_csv("../datafrom200/players.csv")
x_list = player.columns[7:-5]
x_list = ['Finishing', 'Short_Passing', 'Dribbling', 'Long_Passing',
       'Ball_Control', 'Acceleration', 'Sprint_Speed', 'Reactions', 'Stamina',
       'Strength', 'Aggression', 'Penalties', 'Composure',
       'Defensive_Awareness', 'Standing_Tackle', 'Sliding_Tackle']
X = player[x_list]
y = player["value"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
# 计算相关系数
corr_matrix = X.corr()

# 画热力图
plt.figure(figsize=[8, 8])
ax = sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, xticklabels=True, yticklabels=True)
plt.tight_layout()
# plt.savefig("../result/heat_map.pdf")
```

```python
from sklearn.feature_selection import f_classif

# 假设你的数据存储在X和y变量中，其中X是特征向量，y是目标向量
f_scores, p_values = f_classif(X, y)

# 将每个特征的F-Score值和对应的特征名称一起打印出来
a = []
for i, score in enumerate(f_scores):
    print(X.columns[i], score)
    a.append([X.columns[i], score])

pd.DataFrame(a).to_csv("../result/f_scores.csv")
```
