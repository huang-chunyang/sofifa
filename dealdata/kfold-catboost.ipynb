```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold
import pandas as pd
import numpy as np
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
# 读取player 的数据
player = pd.read_csv("../datafrom200/players.csv")
x_list = player.columns[7:-5]
x_list = ['Crossing', 'Finishing', 'Heading_Accuracy', 'Short_Passing', 'Volleys',
       'Dribbling', 'Curve', 'FK_Accuracy', 'Long_Passing', 'Ball_Control',
       'Acceleration', 'Sprint_Speed', 'Agility', 'Reactions', 'Balance',
       'Shot_Power', 'Jumping', 'Stamina', 'Strength', 'Long_Shots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Defensive_Awareness']
# print(x_list)
X = player[x_list]
y = player["value"]
# len(x_list)
```

```python
# 创建模型
import catboost
model = catboost.CatBoostRegressor()

# 创建交叉验证方法
kfold = KFold(n_splits=10, shuffle=True, random_state=0)

# 使用cross_val_score进行交叉验证
scores = cross_val_score(model, X, y, cv=kfold)

# 输出结果
print('Cross-validation scores: \n', scores)
print('Average cross-validation score: {:.3f}'.format(scores.mean()))

# 保存模型
# pd.DataFrame(scores).to_csv("../result/kfold-catboss-r2.csv")
```

```python
# 定义RMSE评分函数
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# 创建RMSE评分器
rmse_scorer = make_scorer(rmse, greater_is_better=False)
# 使用cross_val_score进行交叉验证
scores = - cross_val_score(model, X, y, cv=kfold, scoring=rmse_scorer)

# 输出结果
print('Cross-validation scores: \n', scores)
print('Average cross-validation score: {:.3f}'.format(scores.mean()))
```

```python
pd.DataFrame(scores).to_csv("../result/kfold-catboss-rmse.csv")
```

```python
# 创建模型
import lightgbm as lgb
leaves, depth, n = 31, 19, 500
model = lgb.LGBMRegressor(num_leaves=leaves, max_depth=depth, n_estimators=n)

# 创建交叉验证方法
kfold = KFold(n_splits=10, shuffle=True, random_state=0)

# 使用cross_val_score进行交叉验证
scores = cross_val_score(model, X, y, cv=kfold)

# 输出结果
print('Cross-validation scores: \n', scores)
print('Average cross-validation score: {:.3f}'.format(scores.mean()))

# 保存模型
pd.DataFrame(scores).to_csv("../result/kfold-lgb-r2.csv")

#--------------------------------------------------------------------

# 定义RMSE评分函数
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# 创建RMSE评分器
rmse_scorer = make_scorer(rmse, greater_is_better=False)
# 使用cross_val_score进行交叉验证
scores = - cross_val_score(model, X, y, cv=kfold, scoring=rmse_scorer)

# 输出结果
print('Cross-validation scores: \n', scores)
print('Average cross-validation score: {:.3f}'.format(scores.mean()))
pd.DataFrame(scores).to_csv("../result/kfold-lgb-rmse.csv")
```
